```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_og = pd.read_csv("games.csv")
print(df_og.head())
```

```python
df = df_og.copy()
df = df[df["turns"] > 5]
print(len(df[df["victory_status"] == "mate"]))

df = df[df["victory_status"] == "mate"]
```

```python
w_ratings = df["white_rating"]
b_ratings = df["black_rating"]


print(f"Total # of Games: {len(w_ratings)}")

```

## Distribution of Games based on rating

```python
plt.scatter(w_ratings, b_ratings, alpha=0.5)
plt.show()
```

```python
df_filter_games = df.copy()
df_filter_games = df_filter_games.loc[(df_filter_games["white_rating"] > 1500 )& (df_filter_games["black_rating"] > 1500)]
print(len(df_filter_games))
```
